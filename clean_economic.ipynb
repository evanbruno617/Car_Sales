```python
import pandas as pd
```

```python
ten_year = pd.read_csv('REAINTRATREARAT10Y copy.csv')
ten_year
```

```python
ten_year['Datetime'] = pd.to_datetime(ten_year['DATE'])

ten_year['REAINTRATREARAT10Y_Prev'] = ten_year['REAINTRATREARAT10Y'].shift(1)

ten_year['Ten_Change'] = (ten_year['REAINTRATREARAT10Y'] - ten_year['REAINTRATREARAT10Y_Prev']) / ten_year['REAINTRATREARAT10Y']
```

```python
ten_year
```

```python
ten_year = ten_year.loc[(ten_year['Datetime'] >= pd.to_datetime('2002-01-02')) & (ten_year['Datetime'] < pd.to_datetime('2023-03-01'))
            
            ]
```

```python
ten_year.to_csv('Ten_Year.csv')
```
