```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import os
from sklearn.metrics import r2_score
import statistics as stat
from sklearn.preprocessing import StandardScaler
data_scaler = StandardScaler()
```

```python
convert = {'JAN': 1,
          'FEB': 2,
          'MAR': 3,
          'APR': 4, 
          'MAY': 5,
          'JUN': 6,
          'JUL': 7,
          'AUG': 8,
          'SEP': 9,
          'OCT': 10,
          'NOV': 11,
          'DEC': 12} 
```

```python
sales = pd.read_csv("Car_sales.csv")

items = sales.columns.tolist()[5:-2]

items.append('Ratio_Used_New_Sales_Change')

items.remove( 'Avg_Used_Previous')

items.remove('Avg_New_Previous')

items.remove('Ratio_Used_New_prev')
```

```python
items
```

```python
testing = ['UNRATE','UNRATE_Prev']

results = {}

for item in items:
    results[item] = {}

for test in testing:
    
    sales = pd.read_csv("Car_sales.csv")

    sales['Num_Month'] = sales['Month '].apply(lambda x: convert[x])
    
    dates = []

    for row, i in sales.iterrows():
        date = f"{i['Year ']}-{i['Num_Month']}-01"

        dates.append(date)

    sales['DATE'] = dates

    ten_year = pd.read_csv('UNRATE.csv')

    final_df = pd.merge(ten_year, sales, on='DATE')

    for item in items:

        X = data_scaler.fit_transform(final_df[test].values.reshape(-1,1))

        y = data_scaler.fit_transform(final_df[item].values.reshape(-1,1))

        model = LinearRegression()

        model.fit(X, y)

        y_pred = model.predict(X)

        results[item][f'Coef_{test}'] = model.coef_[0][0]
        results[item][f'R^2_{test}'] = r2_score(y, y_pred)
        
        X = final_df[test].values.reshape(-1,1)

        y = final_df[item]
        
        model = LinearRegression()

        model.fit(X, y)

        y_pred = model.predict(X)
        
        results[item][f'Coef_NON_{test}'] = model.coef_[0]
        
        results[item][f'Coef_Y_{test}'] = model.intercept_


        # displays coef and R^2

pd.DataFrame.from_dict(results, orient='index')
```

```python
# Median Consumer Price Index (CPI) is a measure of core inflation calculated the Federal Reserve Bank of Cleveland and the Ohio State University. Median CPI was created as a different way to get a 'Core CPI' measure, or a better measure of underlying inflation trends. To calculate the Median CPI, the Cleveland Fed analyzes the median price change of the goods and services published by the BLS. The median price change is the price change that's right in the middle of the long list of all of the price changes. This series excludes 49.5% of the CPI components with the highest and lowest one-month price changes from each tail of the price-change distribution resulting in a Median CPI Inflation Estimate.

testing = ['MEDCPIM158SFRBCLE', 'MEDCPIM158SFRBCLE_Prev']

results = {}

for item in items:
    results[item] = {}

for test in testing:
    
    sales = pd.read_csv("Car_sales.csv")

    sales['Num_Month'] = sales['Month '].apply(lambda x: convert[x])
    
    dates = []

    for row, i in sales.iterrows():
        date = f"{i['Year ']}-{i['Num_Month']}-01"

        dates.append(date)

    sales['DATE'] = dates

    ten_year = pd.read_csv('MEDCPIM158SFRBCLE.csv')

    final_df = pd.merge(ten_year, sales, on='DATE')

    for item in items:

        X = data_scaler.fit_transform(final_df[test].values.reshape(-1,1))

        y = data_scaler.fit_transform(final_df[item].values.reshape(-1,1))

        model = LinearRegression()

        model.fit(X, y)

        y_pred = model.predict(X)

        results[item][f'Coef_{test}'] = model.coef_[0][0]
        results[item][f'R^2_{test}'] = r2_score(y, y_pred)

        # displays coef and R^2
        
        X = final_df[test].values.reshape(-1,1)

        y = final_df[item]
        
        model = LinearRegression()

        model.fit(X, y)

        y_pred = model.predict(X)
        
        results[item][f'Coef_NON_{test}'] = model.coef_[0]
        
        results[item][f'Coef_Y_{test}'] = model.intercept_

pd.DataFrame.from_dict(results, orient='index')
```
